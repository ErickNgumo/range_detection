```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

symbols = ['NASDAQ100', 'EURUSD']
timeframes = ['1m', '5m', '15m', '1h']

data_dir = "data/" #folder containing raw CSVs
output_dir = "labeled_data/" #folder for saving saving labeled CSVs
os.makedirs(output_dir, exist_ok=True)

window = 50 #rolling window size
k = 0.5 #threshold multiplier
```

```python
def label_regimes(df, window, k):
    df = df.sort_index()

    tr = pd.concat([
        df['high'] - df['low'],
        (df['high']-df['close'].shift()).abs(),
        (df['low'] - df['close'].shift()).abs()
    ], axis=1).max(axis=1)
    # print(tr)

    df['ATR'] = tr.rolling(window=window).mean()
    df['delta'] = df['close'] - df['close'].shift(window)
    df['vol_scaled'] = df['ATR'] * np.sqrt(window)

    def regime(row):
        if row['delta'] > k* row['vol_scaled']:
            return 'trend_up'
        elif row['delta'] < -k * row['vol_scaled']:
            return 'trend_down'
        else:
            return 'range'
    df['regime'] = df.apply(regime, axis=1)
    return df
```

```python
results = {}

for symbol in symbols:
    for tf in timeframes:
        input_file = f"{data_dir}{symbol}_{tf}.csv"
        output_file = f"{output_dir}{symbol}_{tf}_labeled.csv"

        print(f"\n=== Processing {symbol} {tf} ===")
        try:
            df = pd.read_csv(input_file, parse_dates=['timestamp'], index_col='timestamp')
            df = label_regimes(df, window, k)
            df.to_csv(output_file)

            counts = df['regime'].value_counts(normalize=True).rename("proportion")
            mean_delta = df.groupby('regime')['delta'].mean()

            results[(symbol,tf)] = {'df':df, 'counts': counts, 'mean_delta': mean_delta}

            print(counts)
            print(mean_delta)

            #visualization
            sample = df.tail(2000)
            colors = {'trend_up': 'green', 'trend_down': 'red', 'range': 'gray'}

            plt.figure(figsize=(15,6))
            plt.scatter(sample.index, sample['close'],
                        c=sample['regime'].map(colors), s=8, alpha=0.7)
            plt.plot(sample['close'], color='black', linewidth=0.5)
            plt.title(f"{symbol} ({tf}) with Ground Truth Regimes")
            plt.xlabel("Time")
            plt.ylabel("Price")
            plt.show()
        
        except Exception as e:
            print(f"Error processing {symbol} {tf}: {e}")
```

```python
summary = []
for (symbol, tf), info in results.items():
    row = {
        'symbol': symbol,
        'timeframe': tf,
        **info['counts'].to_dict()
    }
    summary.append(row)

summary_df = pd.DataFrame(summary).fillna(0)
display(summary_df)
```

```python
import pandas as pd
import numpy as np

# ---------------------------
# Parameters
# ---------------------------
rsi_period = 14
ma_short = 10
ma_long = 50

# Example: NASDAQ100 1h labeled dataset
df = pd.read_csv("labeled_data/EURUSD_1h_labeled.csv", parse_dates=['timestamp'], index_col='timestamp')

# ---------------------------
# Compute RSI
# ---------------------------
delta = df['close'].diff()
gain = delta.clip(lower=0)
loss = -delta.clip(upper=0)

avg_gain = gain.rolling(rsi_period).mean()
avg_loss = loss.rolling(rsi_period).mean()

rs = avg_gain / avg_loss
df['RSI'] = 100 - (100 / (1 + rs))

# ---------------------------
# Compute Moving Averages
# ---------------------------
df['MA_short'] = df['close'].rolling(ma_short).mean()
df['MA_long'] = df['close'].rolling(ma_long).mean()
df['MA_slope'] = df['MA_short'] - df['MA_long']

# ---------------------------
# Indicator-based regime predictions
# ---------------------------
def predict_rsi(row):
    if row['RSI'] > 60:
        return 'trend_up'
    elif row['RSI'] < 40:
        return 'trend_down'
    else:
        return 'range'

def predict_ma_slope(row):
    if row['MA_slope'] > 0:
        return 'trend_up'
    elif row['MA_slope'] < 0:
        return 'trend_down'
    else:
        return 'range'

df['RSI_pred'] = df.apply(predict_rsi, axis=1)
df['MA_pred'] = df.apply(predict_ma_slope, axis=1)

# ---------------------------
# Compare with ground truth
# ---------------------------
def accuracy(pred, truth):
    return (pred == truth).mean()

rsi_acc = accuracy(df['RSI_pred'], df['regime'])
ma_acc = accuracy(df['MA_pred'], df['regime'])

print(f"RSI accuracy vs ground truth: {rsi_acc:.3f}")
print(f"MA slope accuracy vs ground truth: {ma_acc:.3f}")

# Optional: confusion matrices
print("\nRSI Confusion Matrix:")
print(pd.crosstab(df['RSI_pred'], df['regime']))

print("\nMA Slope Confusion Matrix:")
print(pd.crosstab(df['MA_pred'], df['regime']))
```

```python
import pandas as pd
import numpy as np
import ta
import os
from sklearn.linear_model import LinearRegression

symbols = ['NASDAQ100', 'EURUSD']
timeframes = ['1m', '5m', '15m', '1h']
labeled_dir = "labeled_data/"

results = []

def rolling_r2_numba(series, window=80):
    """
    Ultra-fast version using numba (requires: pip install numba)
    This can be 10-100x faster for large datasets
    """
    from numba import jit
    
    @jit(nopython=True)
    def _rolling_r2_core(values, window):
        n = len(values)
        result = np.full(n, np.nan)
        
        x = np.arange(window, dtype=np.float64)
        x_mean = x.mean()
        x_centered = x - x_mean
        x_ss = np.sum(x_centered**2)
        
        for i in range(window, n):
            y = values[i-window:i]
            y_mean = y.mean()
            y_centered = y - y_mean
            y_ss = np.sum(y_centered**2)
            
            if y_ss == 0:
                result[i] = 0
                continue
            
            numerator = np.sum(x_centered * y_centered)
            r = numerator / np.sqrt(x_ss * y_ss)
            result[i] = r * r
        
        return result
    
    values = series.values.astype(np.float64)
    result = _rolling_r2_core(values, window)
    return pd.Series(result, index=series.index)

for symbol in symbols:
    for tf in timeframes:
        file = f"{labeled_dir}{symbol}_{tf}_labeled.csv"
        if not os.path.exists(file):
            print(f"Missing: {file}")
            continue
        
        df = pd.read_csv(file, parse_dates=['timestamp'], index_col='timestamp')

        # ========== INDICATORS ==========
        # RSI
        df['rsi'] = ta.momentum.RSIIndicator(df['close'], window=14).rsi()

        # MA Slope
        df['ma_short'] = df['close'].rolling(10).mean()
        df['ma_long'] = df['close'].rolling(50).mean()
        df['ma_slope'] = df['ma_short'] - df['ma_long']

        # ADX
        adx = ta.trend.ADXIndicator(df['high'], df['low'], df['close'], window=14)
        df['adx'] = adx.adx()

        # Bollinger Band Width
        bb = ta.volatility.BollingerBands(df['close'], window=50, window_dev=2)
        df['bb_width'] = (bb.bollinger_hband() - bb.bollinger_lband()) / df['close']

        # R² of trend over rolling window
        # df['r2'] = rolling_r2(df['close'], window=50)
        # df['r2'] = rolling_r2(df['close'], window=50)
        # df['r2'] = np.nan  # placeholder
        df['r2'] = rolling_r2_numba(df['close'], window=50)


        # VWAP and its Gradient
        df['vwap'] = (df['close'] * df['volume']).cumsum() / df['volume'].cumsum()
        df['vwap_grad'] = df['vwap'].diff()

        # Range Ratio (current range vs avg range)
        df['range'] = df['high'] - df['low']
        df['avg_range'] = df['range'].rolling(50).mean()
        df['range_ratio'] = df['range'] / df['avg_range']

        # ========== PREDICTIONS ==========
        def predict_rsi(r):
            if r > 65: return 'trend_up'
            elif r < 35: return 'trend_down'
            return 'range'

        def predict_ma(r):
            if r > 0: return 'trend_up'
            elif r < 0: return 'trend_down'
            return 'range'

        def predict_adx(r):
            if r > 25: return 'trend_up'
            elif r < 20: return 'range'
            return 'trend_down'

        def predict_bb(r):
            if r > 0.015: return 'trend_up'
            else: return 'range'

        def predict_r2(r):
            if r > 0.6: return 'trend_up'
            elif r < 0.4: return 'trend_down'
            return 'range'

        def predict_vwap(g):
            if g > 0: return 'trend_up'
            elif g < 0: return 'trend_down'
            return 'range'

        def predict_range_ratio(r):
            if r < 0.8: return 'range'
            elif r > 1.2: return 'trend_up'
            else: return 'trend_down'

        df['RSI_pred'] = df['rsi'].apply(predict_rsi)
        df['MA_pred'] = df['ma_slope'].apply(predict_ma)
        df['ADX_pred'] = df['adx'].apply(predict_adx)
        df['BB_pred'] = df['bb_width'].apply(predict_bb)
        df['R2_pred'] = df['r2'].apply(predict_r2)
        df['VWAP_pred'] = df['vwap_grad'].apply(predict_vwap)
        df['RR_pred'] = df['range_ratio'].apply(predict_range_ratio)

        # ========== EVALUATE ==========
        def acc(pred):
            return (df[pred] == df['regime']).mean()

        results.append({
            'symbol': symbol,
            'timeframe': tf,
            'RSI_acc': acc('RSI_pred'),
            'MA_acc': acc('MA_pred'),
            'ADX_acc': acc('ADX_pred'),
            'BB_acc': acc('BB_pred'),
            'R2_acc': acc('R2_pred'),
            'VWAP_acc': acc('VWAP_pred'),
            'RangeRatio_acc': acc('RR_pred')
        })

        print(f"{symbol} - {tf}: done")
        

# ========== SUMMARY ==========
summary = pd.DataFrame(results)
print(summary)
```
